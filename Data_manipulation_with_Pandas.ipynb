### **Import wine dataset with sklean**

```python
from sklearn.datasets import load_wine
import pandas as pd

bunch = load_wine()

df = pd.DataFrame(data=bunch['data'], columns=bunch['feature_names'])
df['class'] = bunch['target']
df.head()
```

### **Rename columns name**

```python
df.rename(columns={'class':'target'}, inplace=True)
df.head()
```

### **Group data and count by group**

```python
from sklearn.datasets import fetch_openml

data = fetch_openml('Countries-of-the-World')

df = data.frame
df.head()
```

```python
df.groupby('Region')['Country'].nunique()
```

### **Compute GDP per capita mean by Region**

```python
df.groupby('Region')['GDP_($_per_capita)'].mean()
```

### **Change column type**

```python
df['Population'] = df['Population'].astype('int64')
df['Population'].dtype
```

### **Missing data**

```python
df.isnull().sum()
```

```python
# convert object to float64
df['Deathrate'] = pd.to_numeric(df['Deathrate'], errors='coerce')
df[(df['Deathrate'].isnull())]['Deathrate']
```

### **Fill by Mean**

```python
df['Deathrate'].fillna(df['Deathrate'].mean(), inplace=True)
df['Deathrate'].isnull().sum()
```

### **Fill by zero**

```python
df['Deathrate'].fillna(0, inplace=True)
```

### **Drop all nas**

```python
df.dropna(inplace=True)
```

### **Drop all nas based on a subset**

```python
df.dropna(subset=['Deathrate'], inplace=True)
```

### **Drop lines only if na is present in all columns**

```python
df.dropna(how='all', inplace=True)
```

### **Max, Min, Largest, Smallest**
