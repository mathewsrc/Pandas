### **Import dataset**

```python
import pandas as pd

df = pd.util.testing.makeTimeDataFrame()
df.head()
```

```python
df['datetime'] = df.index
# convert to int as a real example
df['datetime'] = df['datetime'].astype('int64')
df.head()
```

### **Convert not datetime values to datetime**

```python

df['datetime'] = pd.to_datetime(df['datetime'])
df.head()
```

### **Group by year**

```python
df.groupby(df['datetime'].dt.year)['A'].sum()
```

### **Create year, month, day column**

```python
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['day'] = df['datetime'].dt.day
df.head()
```

### **Calculate days of difference**

```python
df['daysOfDiferrence'] = df['day'] - df['day'].min()
df.head()
```

### **Filter by date**

```python
df.loc[(df['datetime'].dt.year == 2000) & (df['datetime'].dt.day == 10)]
```
